# Visualization

```python
sns.pairplot(iris,hue='species',height = 2.5,aspect = 1) #width = height*aspect
#distribution plots
```

```python
iris.hist(edgecolor='black', linewidth = 1.2, figsize=(12,8)) #combines all three species together
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.subplot(2,2,1)
sns.violinplot(x = 'species', y='sepal_length', data = iris)
plt.subplot(2,2,2)
sns.violinplot(x = 'species', y='sepal_width', data = iris)
plt.subplot(2,2,3)
sns.violinplot(x = 'species', y='petal_length', data = iris)
plt.subplot(2,2,4)
sns.violinplot(x = 'species', y='petal_width', data = iris)

#the white dot is the mean
#then the 25%
#the 75%
#the max, the min, and the distribution itself
```

```python
iris.boxplot(by='species',figsize=(12,8))
#green line is the mean
#the box is the 25% and 75%
#group by species
```

```python
pd.plotting.scatter_matrix(iris,figsize=(12,10))
plt.show()
```

```python
sns.pairplot(iris, hue='species', diag_kind='kde')
```
